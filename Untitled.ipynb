```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
import pandas as pd
df = pd.read_csv('new.csv')
df.head()
```

```python
df=df.replace('?', np.nan)
df
```

```python
df=df.dropna()
df
```

```python
df=df.drop(['sym', 'num-of-doors','drive-wheels','wheels-base','length','width','height','curb-weight','num-of-cylinders','fuel-system','bore','stroke','city-mpg','highway-mpg'], axis=1)
```

### OHE

```python
fuel = list(df['fuel-type'].unique())
num = list(range(len(fuel)))
# print(flowers)
# print(num)
mapping = dict(zip(fuel,num))
df['fuel-type']= df['fuel-type'].apply(lambda x: mapping[x])
```

```python
fuel = list(df['aspiration'].unique())
num = list(range(len(fuel)))
# print(flowers)
# print(num)
mapping = dict(zip(fuel,num))
df['aspiration']= df['aspiration'].apply(lambda x: mapping[x])
```

```python
fuel = list(df['body-style'].unique())
num = list(range(len(fuel)))
# print(flowers)
# print(num)
mapping = dict(zip(fuel,num))
df['body-style']= df['body-style'].apply(lambda x: mapping[x])
```

```python
fuel = list(df['engine-location'].unique())
num = list(range(len(fuel)))
# print(flowers)
# print(num)
mapping = dict(zip(fuel,num))
df['engine-location']= df['engine-location'].apply(lambda x: mapping[x])


fuel = list(df['engine-type'].unique())
num = list(range(len(fuel)))
# print(flowers)
# print(num)
mapping = dict(zip(fuel,num))
df['engine-type']= df['engine-type'].apply(lambda x: mapping[x])

fuel = list(df['make'].unique())
num = list(range(len(fuel)))
# print(flowers)
# print(num)
mapping = dict(zip(fuel,num))
df['make']= df['make'].apply(lambda x: mapping[x])
```
